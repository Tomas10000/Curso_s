```python
import pandas as pd

# Suponiendo que df_joined_4 es tu DataFrame con la columna 'amount'

# Calcular el primer cuartil (Q1) y el tercer cuartil (Q3)
Q1 = df_joined_4['amount'].quantile(0.25)
Q3 = df_joined_4['amount'].quantile(0.75)

# Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# Calcular los límites inferior y superior
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Marcar los valores atípicos en la columna 'amount'
df_joined_4['outlier'] = 'no_outlier'  # Asignar por defecto que no son outliers

# Marcar los valores atípicos bajos y altos
df_joined_4.loc[df_joined_4['amount'] < lower_limit, 'outlier'] = 'outlier_bajo'
df_joined_4.loc[df_joined_4['amount'] > upper_limit, 'outlier'] = 'outlier_alto'

# Mostrar el DataFrame con la nueva columna 'outlier' que marca los valores atípicos
df_joined_4

```

```python
import pandas as pd

# Suponiendo que df_joined_4 es tu DataFrame con las columnas 'outlier' y 'date'

# Asegurarse de que la columna 'date' esté en formato datetime
df_joined_4['date'] = pd.to_datetime(df_joined_4['date'])

# Extraer la hora de la columna 'date'
df_joined_4['hour'] = df_joined_4['date'].dt.hour + df_joined_4['date'].dt.minute / 60.0

# Crear la columna 'target' con la condición que mencionas
df_joined_4['target'] = ((df_joined_4['outlier'] == 'outlier_alto') & 
                          (df_joined_4['hour'] >= 6) & 
                          (df_joined_4['hour'] < 8)).astype(int)

# Mostrar el DataFrame con la nueva columna 'target'
df_joined_4
```

```python
df_joined_5 = df_joined_4.drop(columns=['outlier', 'amount_2', 'outlier', 'target', 'name_merchant', 'card', 'id_card_holder','date','id_transaction','name_category'])
df_joined_5
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import joblib

# Suponiendo que df_joined_5 es tu DataFrame

# Definir las características (X) y la variable objetivo (y)
X = df_joined_5[['amount', 'id_merchant', 'id_merchant_category', 'hour']]  # Características
y = df_joined_5['y']  # Variable objetivo

# Dividir los datos en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de RandomForest
model = RandomForestClassifier(n_estimators=10, random_state=11)
model.fit(X_train, y_train)

# Hacer predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)

# Mostrar el reporte de clasificación (precisión, recall, F1, etc.)
print("Reporte de clasificación:\n")
print(classification_report(y_test, y_pred))

# Calcular y mostrar las métricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Guardar el modelo entrenado en formato .pkl
joblib.dump(model, 'modelo_entrenado.pkl')

print("Modelo guardado con éxito en 'modelo_entrenado.pkl'.")
```
